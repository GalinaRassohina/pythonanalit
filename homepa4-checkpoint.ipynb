#  3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = np.round(df.corr())
corr_matrix
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(corr_matrix, annot = True, linewidths=5,cmap = 'coolwarm');
```

```python
plt.figure(figsize=(8,4))
sns.lmplot(data=df, x="sqft_living", y="price")
plt.xlabel('Жилая лощадь')
plt.ylabel('Стоимость дома')
plt.title('Стоимости дома от  жилой площади');
```

Вывод: Наиболее покупаемая недвидимость квадратура менее 6000 кв.м. Чем большк площадь жилья тем ниже спрос

```python
plt.figure(figsize=(8,4))
sns.boxplot(data= df, x ='sqft_living', y= 'price')
plt.xlabel('Класс дома')
plt.ylabel('Стоимость дома')
plt.title('Зависимость стоимости дома от его жилой площади ');
```

```python
plt.figure(figsize=(8, 8))
sns.violinplot(data = df, x = 'grade', y = 'price')
plt.xlabel('Класс дома')
plt.ylabel('Стоимость дома')
plt.title('Зависимость стоимости дома от его класса');
```

Вывод: Дома имеющие класс от 3 до 6 приобретают примерно с одинаковой частотой

```python
plt.figure(figsize=(11, 6))
sns.boxplot(data = df, x = 'waterfront', y = 'price')
plt.xlabel('Расположение на береговой линии')
plt.ylabel('Стоимость ')
plt.title('Распределение стоимости домов зависимости от располоения на береговой линии');
```

Вывод Стоимость на дома находящихся береговой линии значительно выше, чем на дома находяшиеся вдалии от береговой линии.
При этом цены на жилье на береговой линии имеют более планвный переход. Стоимость же жилья вне береговой линии
меняется скачкообразно (т.е. встречается недвижимость с ценами выше чем на береговой линии и давольно частый случай).

```python
plt.figure(figsize=(8,6))
sns.jointplot(data=df, x="sqft_above", y="price")
plt.xlabel('Площадь застройки жилой площади')
plt.ylabel('Стоимость дома')
plt.title('Зависимость стоимости от площади застройки', loc='left');
```

Вывод: Чем больше площадь застройке , тем меньше спрос.Намболее популярная площадь от 2000 кв.м до 4000 кв.м

```python
plt.figure(figsize=(12,6))
sns.histplot(data = df, x = 'bedrooms', y = 'price', bins = 50)
plt.xlabel('Количество спален')
plt.ylabel('Стоимость дома')
plt.title('Распределение стоимости домов в зависимости от количества спален');
```

Вывод: Самые восстребуемые дома с 5 спальнями. Спрос до 5 спален растет, а после 5 спален, начинает снижаться.
